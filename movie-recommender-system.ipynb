```python
# Importing pandas and numpy
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
# Importing datasets
movies = pd.read_csv(r'C:\Users\Donlyone7\OneDrive\Desktop\Movie Reccomneding System\Datasets\tmdb_5000_movies.csv')
credits = pd.read_csv(r'C:\Users\Donlyone7\OneDrive\Desktop\Movie Reccomneding System\Datasets\tmdb_5000_credits.csv')
```

```python
# Merging 'movies' and 'credits' datasets into 'movies'
movies = movies.merge(credits,on='title')
```

```python
movies.info()
# Columns to keep:
# id
# title
# overview
# genres
# cast
# crew
# keywords
```

```python
# Taking the columns that we need and keeping only those in 'movies'
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
# Cheking again for null values
movies.isnull().sum()
```

```python
# Checking for duplicates
movies.duplicated().sum()
```

```python
# Dropping null values
movies.dropna(inplace=True)
```

```python
# Creating and storing new form factors of the columns
```

```python
movies['genres'] = movies['genres'].apply(convert) # genres
movies['keywords'] = movies['keywords'].apply(convert) # keywords
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert) # cast
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
# We only need director names, because that's the only thing that everybody care!
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) # crew
movies.head()
```

```python
movies.sample(5)
```

```python
# Removing spaces so that after we convert one name into tags their first name and last name doesn't become two separate tags.
# Sample applies for genres too.
# E.g, 'Christian Bale' = 'ChristianBale'.

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# Applying the function to the columns
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
# Checking
movies.head()
```

```python
# Coverting the 'overview' column into a list and storing it
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Creating a new column 'tags' by concatenateing 'overview', 'cast', 'crew', 'keywords'
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# Dropping the columns that we don't need
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
# Converting our 'tags' column into a string
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
# Importing CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
# Checking our vectors
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Storing the variables
similarity = cosine_similarity(vector)
```

```python
# Checking
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Interstellar')
```
